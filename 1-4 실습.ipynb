```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

major_xticks = [i for i in range(0, 101, 20)]
minor_xticks = [i for i in range(0, 101, 5)]

major_yticks = [i for i in range(0, 11, 2)]
minor_yticks = [i for i in range(0, 11, 1)]

ax.set_xticks(major_xticks)
ax.set_xticks(minor_xticks, minor=True)

ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks, minor=True)

ax.tick_params(axis='x', labelsize=15, length=7, width=3)
ax.tick_params(axis='x', which='minor', length=3, width=1)

ax.tick_params(axis='y', rotation=90, labelsize=15, length=7, width=3)
ax.tick_params(axis='y', which='minor', length=3, width=1)
```

```python
xticks = [i for i in range(10)]
xtick_labels = ['Class ' + str(i) for i in xticks]

yticks = [i for i in range(0, 301, 30)]
ytick_labels = [str(i) + ' points' for i in yticks]

minor_yticks = [i for i in range(15, 301, 30)]

fig, ax = plt.subplots(figsize=(7, 7))

ax.set_xticks(xticks)
ax.set_xticklabels(xtick_labels)
ax.set_yticks(yticks)
ax.set_yticklabels(ytick_labels)

ax.set_yticks(minor_yticks, minor=True)

ax.tick_params(axis='x', rotation=90, labelsize=15, direction='in')
ax.tick_params(axis='y', labelsize=15, direction='in', width=3, length=6)
ax.tick_params(axis='y', which='minor', direction='in', width=2, length=2)


ax.set_xlabel('class', fontsize=25, labelpad=10)
ax.set_ylabel('point number', fontsize=25, labelpad=10)

fig.subplots_adjust(bottom=0.1, left=0.1)
```
