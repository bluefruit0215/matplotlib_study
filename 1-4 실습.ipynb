```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

major_xticks = [i for i in range(0, 101, 20)]
minor_xticks = [i for i in range(0, 101, 5)]

major_yticks = [i for i in range(0, 11, 2)]
minor_yticks = [i for i in range(0, 11, 1)]

ax.set_xticks(major_xticks)
ax.set_xticks(minor_xticks, minor=True)

ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks, minor=True)

ax.tick_params(axis='x', labelsize=15, length=7, width=3)
ax.tick_params(axis='x', which='minor', length=3, width=1)

ax.tick_params(axis='y', rotation=90, labelsize=15, length=7, width=3)
ax.tick_params(axis='y', which='minor', length=3, width=1)
```
